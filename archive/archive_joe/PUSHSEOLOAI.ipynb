```python
def convertDoc(doc):
    json = {}
    keyword = doc["keyword"]
    city = f"{doc['city']}"
    state = f"{doc['state']}"
    slug = doc["slug"].replace("-services","")
    short = slug.split('-')[-1]
    json["cityName"] = city
    json["stateName"] = state
    json["category"] = city
    json["keyword"] = keyword
    json["metaTitle"] =  f'Best {keyword} {city} {short.upper()} | We Revolutionize {keyword} For {city} Starups'
    json["metaDescription"] = f"AI is redefining {keyword} Copywriting services in {city}, {state}. Automated Content Writing Plugin(s) For Wordpress & Strapi. Your new {keyword} {city} expert is here. Free SEO Strategy Session. Book Today."
    # json["metaTitle"] =  doc["web"]["pageTitle"].replace("World-Class", "Best").replace("Exam","")
    if random.uniform(0,1) < 0.5:
        json["metaTitle"] = json["metaTitle"].replace("In", "Near")
        print("Near")
    # f"{doc['web']['metaTitle']}" # f"Best {doc['keyword']} {doc['city']} {doc['slug'].split('-')[-1].upper()}"
    # json["metaDescription"] = doc["web"]["meta"][1]["content"]# f"{doc['web']['metaDescription']}" # f"Best {doc['keyword']} Near {doc['city']}. {doc['state']} qualified {doc['keyword']} experts. Free Quote & Installation. Call Today."
    faq = []
    for f in doc["faq"]:
        val = f["question"].replace("{COMPANY}", "Company")
        f["question"] = f"{str(val)}".replace("\'","’").replace("\n\n", "<br/>").replace("\n", "<br/>").replace(" \""," “").replace("\" ","” ").replace("\".","”.").replace("\",","”,").replace(",\"",",”").replace(";","；").replace("\"","“").replace("|","-").replace("&amp;ndash;","-").replace("\\","").replace("\'","'")
        val = f["answer"].split("Blog post title:")[0].split("Blog Post")[0].split("Post navigation")[0].split("Post Navigation")[0].split("Website Name:")[0].split("Webpage:")[0].split("Visit:")[0].split("Web Page:")[0].split("Blog title:")[0].split("\n\n")[0].split("Find this information")[0].split("Webinars:")[0].split("Site:")[0].split("Sections:")[0].split("\nThis website uses cookies")[0]#.split("1.2M+ Students Traded")[0]
        f["answer"] = val.replace("\'","’").replace("\n\n", "<br/>").replace("\n", "<br/>").replace(" \""," “").replace("\" ","” ").replace("\".","”.").replace("\",","”,").replace(",\"",",”").replace(";","；").replace("\"","“").replace("|","-").replace("&amp;ndash;","-").replace("\\","").replace("\'","'")
        faq.append(f)
    json["slug"] = slug
    print("New ", json["slug"])
    json["faq"] = faq
    return json
```
